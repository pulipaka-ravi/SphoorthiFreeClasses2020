```python
emailurl="https://gist.githubusercontent.com/itsbalamurali/8231301/raw/c603eb4389ea6df78ea74c12f905c664ddabc55e/emails.txt"
```

```python
resp=requests.get(emailurl)
```

```python
data=resp.text
```

```python
data
```

```python
data.split('\n')
```

```python
email_lines=[]
```

```python
for line in data.split('\n'):
    if '@' in line and '.' in line:
        email_lines.append(line)
        
```

```python
email_lines
```

```python
'AS54419:netadmin@ravecinemas.com'.split(":")[1]
```

```python
emails=[]
```

```python
for eachline in email_lines:
    e=eachline.split(":")[1]
    emails.append(e)
```

```python
emails[0]
```

```python
emails[0].split(',')
```

```python
sample=[]
```

```python
sample.extend(emails[0].split(','))
```

```python
sample
```

```python
for eachline in emails:
    if ',' in eachline:
        print(eachline)
```

```python
for eachline in emails:
    if ',' in eachline:
        result=eachline.split(',')
        sample.extend(result)
```

```python
sample
```

```python
for eachline in emails:
    if ',' in eachline:
        result=eachline.split(',')
        sample.extend(result)
    else:
        sample.append(eachline)
```
